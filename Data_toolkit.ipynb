3. Explain the following terms:

 The difference in nparray, npasarray and npasanyarray

 The difference between Deep copy and shallow copy

1. Differences in np.array, np.asarray, and np.asanyarray

np.array():
 Creates a new NumPy array from a given object (like a list or a tuple).

 np.asarray():
 Converts the input to an array, but does not make a copy if the input is already an array.

 np.asanyarray():
 Similar to np.asarray(), but it allows subclasses of ndarray (like masked arrays) to be returned.

 2. Differences Between Deep Copy and Shallow Copy

Shallow Copy:
 A shallow copy creates a new object, but does not create copies of nested objects. Instead, it copies references to those objects.

 Deep Copy:
 A deep copy creates a new object and recursively copies all nested objects. This means all levels of objects are duplicated.

18. With Bokeh plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function?

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output (use output_file() for HTML file output)
output_notebook()  # Uncomment this line if running in a Jupyter Notebook

# Generate data for the sine wave
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title='Sine Wave Function', x_axis_label='X', y_axis_label='sin(X)',
           width=800, height=400)

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label='sin(x)', line_width=2, color='blue')

# Add grid lines
p.grid.grid_line_alpha = 0.5

# Show the plot
show(p)
```

19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool

# Prepare the output (use output_file() for HTML file output)
output_notebook()  # Uncomment this line if running in a Jupyter Notebook

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame for better data handling
data = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a Bokeh figure
p = figure(x_range=data['categories'], title='Random Categorical Bar Chart',
           toolbar_location=None, tools="")

# Color bars based on values
colors = [f"#{int(255 * (value / max(values))):02x}00{int(255 * (1 - value / max(values))):02x}" for value in values]

# Add bars to the figure
p.vbar(x='categories', top='values', source=source,
       width=0.9, line_color='white', fill_color=colors)

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'
p.y
```

20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
'Simple Line Plot'

```python
import numpy as np
import plotly.graph_objects as go

# Generate random data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.random.rand(100)  # Random values between 0 and 1

# Create a Plotly figure
fig = go.Figure()

# Add a line trace
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Update layout
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis Label',
    yaxis_title='Y-axis Label',
)

# Show the plot
fig.show()
```

21. Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set
the title as 'Interactive Pie Chart'.

```python
import numpy as np
import plotly.graph_objects as go

# Generate random data
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 100, size=len(labels))  # Random values for each category

# Create a Plotly figure for the pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values,
                               textinfo='label+percent',
                               insidetextorientation='radial')])

# Update layout with title
fig.update_layout(title='Interactive Pie Chart')

# Show the plot
fig.show()
```
