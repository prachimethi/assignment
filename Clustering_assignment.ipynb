# **THEORITICAL QUESTIONS**

Ques-1 What is unsupervised learning in the context of machine learning ?

Unsupervised learning is a type of machine learning where the model is trained on data without explicit labels or predefined outcomes. In other words, the algorithm tries to learn the underlying structure or patterns from the data without any guidance on what the correct answers are.

The goal of unsupervised learning is typically to find hidden patterns, groupings, or relationships in the data. This contrasts with supervised learning, where the model is trained on labeled data and learns to map inputs to known outputs or when there is no target variables.

Ques-2 How does K-Means clustering algorithm work ?

The K-Means clustering algorithm is one of the most popular and widely used unsupervised learning techniques for clustering. The idea is to partition a dataset into K distinct clusters based on the similarities between data points, where K is a user-specified number of clusters.

Ques-3  Explain the concept of a dendrogram in hierarchical clustering ?

In hierarchical clustering, a dendrogram is a tree-like diagram that illustrates the arrangement of clusters at different levels of similarity. It visually represents the process of merging (agglomerative approach) or splitting (divisive approach) data points or clusters step by step, showing how clusters are formed as the algorithm proceeds.

Ques-4 What is the main difference between K-Means and Hierarchical Clustering ?

The main difference between K-Means and Hierarchical Clustering lies in their approach to forming clusters. K-Means is a partitional method that requires you to specify the number of clusters (K) beforehand and works by iteratively assigning data points to the nearest centroids and updating the centroids until convergence. It is more efficient for large datasets but assumes spherical, equally sized clusters and can struggle with outliers. In contrast, hierarchical clustering builds a tree-like structure (dendrogram) that doesn't require specifying the number of clusters upfront, offering flexibility in choosing the number of clusters after observing the tree. Hierarchical clustering can handle arbitrary-shaped clusters and is less sensitive to outliers but tends to be computationally more expensive, especially for large datasets. While K-Means is faster and better suited for large datasets, hierarchical clustering provides a more detailed view of the data's structure through the dendrogram.

Ques-5 What are the advantages of DBSCAN over K-Means ?

DBSCAN is particularly useful when you have data with clusters of arbitrary shapes, varying densities, or when outliers are present, as it can find meaningful clusters without being influenced by noise. In contrast, K-Means is better suited for datasets where clusters are roughly spherical, the number of clusters is known beforehand, and outliers are less of a concern.

Ques- When would you use Silhouette Score in clustering ?

The Silhouette Score is a metric used to evaluate the quality of clusters formed by a clustering algorithm. It provides an indication of how well each data point has been clustered, both in terms of how close it is to points within the same cluster and how far it is from points in other clusters.

Ques-7 What are the limitations of Hierarchical Clustering ?

Hierarchical clustering has several limitations, including high computational complexity (O(n²)), which makes it inefficient for large datasets. It can also be sensitive to noise and outliers, as even a few outliers can distort the cluster structure. Additionally, it struggles with non-spherical or unevenly sized clusters and lacks flexibility in directly controlling the number of clusters. The choice of linkage criteria (e.g., single, complete, or average linkage) can significantly affect the results, and the algorithm requires substantial memory to store the distance matrix. These factors make hierarchical clustering less suitable for very large, high-dimensional, or noisy datasets.


Ques-8 Why is feature scaling important in clustering algorithms like K-Means ?

Feature scaling is important in clustering algorithms like K-Means because the algorithm relies on distance metrics (typically Euclidean distance) to assign data points to clusters and update cluster centroids. If the features in the dataset have different scales or units, the clustering results can be heavily biased toward features with larger numerical ranges, which can distort the clustering process.

Ques-9 How does DBSCAN identify noise points ?

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) identifies noise points by analyzing the density of data points in the dataset.

Ques-10 Define inertia in the context of K-Means ?

In the context of K-Means clustering, inertia (also known as within-cluster sum of squares or WSS) is a measure of how well the data points are clustered around their respective centroids. It quantifies the compactness of the clusters, i.e., how close the data points within each cluster are to the centroid of that cluster.

Ques-11  What is the elbow method in K-Means clustering ?

The Elbow Method is a technique used to determine the optimal number of clusters (K) in K-Means clustering. It helps to find the point at which adding more clusters no longer significantly improves the clustering results. The basic idea is to plot the inertia (or the within-cluster sum of squares) against different values of K and look for an "elbow" point in the plot.

Ques-12 Describe the concept of "density" in DBSCAN ?

In DBSCAN, density refers to the concentration of data points within a specified radius (ε). A cluster is defined as a dense region of points, where each point in a cluster is connected to other points in the same cluster by reaching core points that have sufficient nearby points (MinPts). Points that do not meet this density criterion are labeled as noise or outliers. This density-based approach enables DBSCAN to discover clusters of arbitrary shape and efficiently identify noise points in datasets.





Ques-13 Can hierarchical clustering be used on categorical data ?

Yes, hierarchical clustering can be used on categorical data, but it requires modifications in the way distances are calculated. Unlike numerical data, where Euclidean or other distance metrics are directly applicable, categorical data typically involves discrete values, such as labels or categories, which necessitate different distance measures.

Ques-14 What does a negative Silhouette Score indicate ?

A negative Silhouette Score indicates that the data points are likely misclassified or assigned to the wrong cluster. Specifically, it suggests that a data point is closer to points in a different cluster than to points within its own cluster

Ques-15  Explain the term "linkage criteria" in hierarchical clustering ?

In hierarchical clustering, the term "linkage criteria" refers to the method used to define the distance or similarity between two clusters during the process of merging (in agglomerative clustering) or splitting (in divisive clustering). It determines how the distance between two clusters is calculated based on the distances between individual points within those clusters.

Ques-  Why might K-Means clustering perform poorly on data with varying cluster sizes or densities ?

K-Means clustering performs poorly on data with varying cluster sizes or densities because it assumes clusters are spherical and evenly distributed. It uses Euclidean distance, which can misclassify points in regions with different densities, and struggles with non-convex or irregularly shaped clusters. Additionally, K-Means is sensitive to the initial placement of centroids, making it prone to suboptimal solutions, especially when clusters have varying sizes. As a result, K-Means may fail to correctly separate clusters in such cases.

Ques-17 What are the core parameters in DBSCAN, and how do they influence clustering ?

In DBSCAN, the two core parameters are:

1. Epsilon (ε): Defines the radius around a point to consider it as a neighbor. A small ε results in many points being classified as noise, while a large ε may merge distinct clusters.

2. MinPts (Minimum Points): Specifies the minimum number of points required to form a dense cluster. A high MinPts value leads to larger, denser clusters, while a low value may result in smaller, less defined clusters

Ques-18 How does K-Means++ improve upon standard K-Means initialization ?

K-Means++ improves upon standard K-Means by carefully selecting the initial centroids to be well-spread out, leading to better clustering results, faster convergence, and a reduced likelihood of getting stuck in local minima.

Ques-19 What is agglomerative clustering?

Agglomerative Clustering is a type of hierarchical clustering algorithm that follows a bottom-up approach. In this method, each data point starts as its own individual cluster, and the algorithm iteratively merges the closest (most similar) clusters until all points are part of a single cluster or a stopping criterion is met (like a specified number of clusters).

Ques-20 What makes Silhouette Score a better metric than just inertia for model evaluation?

The Silhouette Score is often considered a better metric than inertia (within-cluster sum of squared distances) for model evaluation in clustering because it provides a more comprehensive evaluation of the quality of the clusters, considering both cohesion and separation.

Ques-21 Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot ?

```python
#Import neccessary libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#Devide x and y
x , y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=1)

#apply k means
km = KMeans(n_clusters=4)
km.fit(x)
ypred = km.predict(x)

#visulaization
plt.figure(figsize=(8, 6))
plt.scatter(x[:, 0], x[:, 1], c= ypred, cmap='viridis', s=50)

# Plot the centers
centers = km.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')

plt.title('K-Means Clustering with 4 Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

Ques-28 Load the Digits dataset, reduce it to 2D using PCA, and visualize clusters from K-Means ?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Step 1: Load the Digits dataset
digits = load_digits()
X = digits.data  # The features (1797 samples, 64 features)
y = digits.target  # The target labels (which are the actual digit classes)

# Step 2: Apply PCA to reduce the data to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Step 3: Apply K-Means clustering with 10 clusters (one for each digit)
kmeans = KMeans(n_clusters=10, random_state=42)
y_kmeans = kmeans.fit_predict(X_pca)

# Step 4: Visualize the clustering result
plt.figure(figsize=(8, 6))

# Plot the data points with different colors for each cluster
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='viridis', s=50, alpha=0.7)

# Plot the centroids of the clusters
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X', label='Centroids')

plt.title('K-Means Clustering on Digits Dataset (Reduced to 2D using PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

Ques-29 Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart ?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Step 1: Generate synthetic data using make_blobs
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# Step 2: Evaluate silhouette scores for k = 2 to 5
silhouette_scores = []

for k in range(2, 6):
    # Apply K-Means clustering
    kmeans = KMeans(n_clusters=k, random_state=42)
    y_kmeans = kmeans.fit_predict(X)

    # Calculate the silhouette score
    score = silhouette_score(X, y_kmeans)
    silhouette_scores.append(score)

# Step 3: Display the silhouette scores as a bar chart
plt.figure(figsize=(8, 6))
plt.bar(range(2, 6), silhouette_scores, color='skyblue', edgecolor='black')

# Adding titles and labels
plt.title('Silhouette Scores for K-Means Clustering (k=2 to 5)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(range(2, 6))  # Set x-ticks for k=2 to 5
plt.show()
```

Ques-31 Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries?

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Generate synthetic data with overlapping clusters
X, y = make_blobs(n_samples=300, centers=3, cluster_std=2.0, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)

# Plot decision boundaries and clusters
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

# Predict on the meshgrid points
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Create the plot
plt.figure(figsize=(10, 8))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans, s=50, cmap='coolwarm', edgecolors='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='X', s=200, c='black', label='Centroids')
plt.title('K-Means Clustering with Decision Boundaries')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

Ques-47 Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
count ?

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# Generate synthetic data with 3 centers and noise
X, _ = make_blobs(n_samples=1000, centers=3, cluster_std=0.60, random_state=42)

# Apply DBSCAN to identify clusters and noise
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Count the number of clusters and noise points
# Noise points are labeled as -1 by DBSCAN
num_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
num_noise_points = np.sum(dbscan_labels == -1)

# Print the results
print(f"Number of clusters: {num_clusters}")
print(f"Number of noise points: {num_noise_points}")
```

Ques-48 Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the
clusters ?

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns
import pandas as pd

# Load the Digits dataset
digits = load_digits()
X = digits.data
y = digits.target

# Reduce dimensions using t-SNE to 2D
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Apply Agglomerative Clustering
agg_clust = AgglomerativeClustering(n_clusters=10)  # 10 clusters since there are 10 digits
labels = agg_clust.fit_predict(X_tsne)

# Create a DataFrame for plotting
df = pd.DataFrame(X_tsne, columns=['TSNE-1', 'TSNE-2'])
df['Cluster'] = labels

# Plot the clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x='TSNE-1', y='TSNE-2', hue='Cluster', data=df, palette='viridis', s=60, marker='o', legend=None)
plt.title("Agglomerative Clustering on t-SNE-reduced Digits Data")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```
